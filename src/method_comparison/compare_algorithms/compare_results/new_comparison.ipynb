```r
plot_nr_samples <- function(my_df
                           ,name){
    options(repr.plot.height = 5
        ,repr.plot.width = 5)
plot <- ggplot(data = my_df
      ,aes(x = number_healthy
          ,y = number_aml
          )
      )+
geom_jitter()+ 
theme_bw()+
xlab("number of healthy samples")+
ylab("number of AML samples")+
theme(text = element_text(size=22)
      ,plot.title = element_text(margin = margin(t = 0, r = 0, b = 15, l = 0))
      ,axis.title.y = element_text(margin = margin(t = 0, r = 15, b = 0, l = 0))
     )+
ggtitle(name)
    ggsave(file=paste0("outs/", name, "_nr_samples.jpeg"), plot=plot, width=10, height=8)

    plot
}
```

```r
extract_pairs <- function(x) {
  # Split the string by '_'
  parts <- unlist(strsplit(x, '_'))
  # Split each part by ':' and take the second element
  ligand <- unlist(strsplit(parts[1], ':'))[2]
  receptor <- unlist(strsplit(parts[2], ':'))[2]
  # Combine as ligand_receptor
  paste(ligand, receptor, sep = '_')
}
```

```r
uniq_pairs = sapply(aml_uniq, extract_pairs)
```
