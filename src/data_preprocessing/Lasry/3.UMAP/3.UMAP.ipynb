### load data

````python
import numpy as np
import scanpy as sc
import pandas as pd
import os
import shutil
````

````python
print("here")
````

````python
adata=sc.read_csv("lasry_batch_corr.csv.gz")
````

````python
adata = adata.T
````

````python
# read in cell annotation file
anno_cells = pd.read_csv("anno_cells_batch_corr.txt", sep = "\t") 
````

````python
# read in sample annotation file
anno_samples = pd.read_csv("anno_samples_batch_corr.txt", sep = "\t") 
````

````python
# read in gene annotation file
anno_genes = pd.read_csv("anno_genes_batch_corr.txt", sep = "\t") 
````

````python
# load cell subtype and color keys
cell_relabel=pd.read_csv("../../cell_relabelling.csv"
                        ,sep=';')
````

````python
cell_relabel = cell_relabel[cell_relabel.source=="Lasry et al., 2022"]
````

### process data

````python
adata.var_names=anno_genes['gene_symbol']
print("adata.var_names[1:10]")
print(adata.var_names[1:10])
adata.var = anno_genes
adata.var.index = adata.var.index.astype(str)
````

````python
adata.obs_names=anno_cells['cell_ID']
print("adata.obs_names[1:10]")
print(adata.obs_names[1:10])
adata.obs = anno_cells
adata.obs.index = adata.obs.index.astype(str)
````

### prepare for the visualization

````python
# define colors
colors_cell_type=dict(zip(cell_relabel["cell_type"],cell_relabel["cell_type_color_hex"]))
print(colors_cell_type)
````

````python
#there is an error at line 9, cell_subtype is only 2 values?
adata.obs["cell_subtype"].value_counts()
````

````python
# rename "malignangt" columns into "bares_mutatoin"
adata.obs["bares_mutation"] = adata.obs["malignant"].astype('str')
````

````python
# define color schemes
colors_cell_subtype=dict(zip(cell_relabel["cell_subtype"],cell_relabel["cell_subtype_color_hex"]))
print(colors_cell_subtype)

colors_cell_type=dict(zip(cell_relabel["cell_type"],cell_relabel["cell_type_color_hex"]))
print(colors_cell_type)
````

````python
adata.uns["health_status_colors"] = ["#7C001F" # bordeau for AML
                                                     , "#7ac5cd" # CadetBlue3 for healthy
                                                     ]

adata.uns["bares_mutation_colors"] = ["#A6ACAF" # grey for FALSE
                                                  , "#C0392B" # red for TRUE
                                                 ]
````

### visualize before batch correction

````python
# logtransform before HVG calculation
sc.pp.log1p(adata)

sc.pp.highly_variable_genes(adata)
sc.pl.highly_variable_genes(adata)
````

````python
# print how many HVGs we have:
print("Nr of HVGs:")
print(sum(adata.var.highly_variable))
````

````python
sc.tl.pca(adata, svd_solver='arpack')
````

````python
sc.pl.pca(adata
          , color=["cell_type"]
          ,palette=colors_cell_type
          ,save="_adata_cell_type.pdf"
         )
````

````python
sc.pp.neighbors(adata)
````

````python
sc.tl.umap(adata)
````

````python
sc.pl.umap(adata
           , color=["cell_type"]
           ,palette=colors_cell_type
           ,save="_adata_cell_type.pdf"
          )
````

````python
sc.pl.umap(adata
           , color=["cell_subtype"]
          ,palette=colors_cell_subtype
          ,save="_adata_cell_subtype.pdf"
          )
````

````python
sc.pl.umap(adata
           , color=["sample_ID"]
          ,save="_adata_sample_ID.pdf"
          )
````

````python
sc.pl.umap(adata
           , color=["health_status"]
          ,save="_adata_health_status.pdf"
          )
````

````python
sc.pl.umap(adata
           , color=["bares_mutation"]
          ,save="_adata_malignant.pdf"
          )
````

````python
sc.pl.umap(adata
           , color=["cell_type_original"]
           #,legend_loc='center'
          ,save="_adata_cell_type_original.pdf"
          )
````

````python
# export counts as csv.gz
print("save counts_corr.csv.gz")

counts_corr=adata.to_df().transpose()
counts_corr.index = adata.var.gene_symbol
counts_corr.columns = adata.obs.cell
counts_corr.to_csv(output_path + "counts_corr.csv.gz"
                   ,index=True
                   ,compression="gzip"
                   )
````

````python
# export anno_cells_corr
print("save anno_cells_corr.txt")
adata.obs.to_csv(output_path + "anno_cells_corr.txt"
                    ,sep = "\t"
                    ,index = True)
````

````python
# export anno_samples_corr
print("save anno_samples_corr.txt")
anno_samples.to_csv(output_path + "anno_samples_corr.txt"
                    ,sep = "\t"
                    ,index = True)
````

````python
# export anno_genes_corr
print("save anno_genes_corr.txt")
adata.var.to_csv(output_path + "anno_genes_corr.txt"
                    ,sep = "\t"
                    ,index = True)
````
